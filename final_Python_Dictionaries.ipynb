##  What is a Dictionary?

A **dictionary** in Python is a **data structure** that stores data in the form of **key : value** pairs.

üëâ It means:

* Each piece of data has a **name (key)**
* And the **actual data (value)**

### Example:

```python
"name" : "Anisha"
```

Here:

* `"name"` ‚Üí key
* `"Anisha"` ‚Üí value


## Why Do We Use a Dictionary?

Let‚Äôs understand this with a real example.

### ‚ùå Without Dictionary (List)

```python
marks = [80, 90, 85]
```

### ‚úÖ With Dictionary

```python
marks = {
    "Maths": 80,
    "Science": 90,
    "English": 85
}
```

Now it‚Äôs clear:

* Maths ‚Üí 80
* Science ‚Üí 90

üìå **This clarity is the main reason we use dictionaries.**

##  Characteristics of Dictionary

A dictionary has the following characteristics:

1. Stores data as **key : value**
2. **Keys must be unique**
3. **Keys must be immutable**
4. **Values can be mutable or immutable**
5. Dictionary is **mutable** (can be changed)
6. No index (not accessed by position)
7. Uses **hashing** ‚Üí very fast access

# Understanding Key and Value (Very Important)
## What is a KEY? üîë



A **key** is a **name or identifier** used to access data.

Think of:

* Phone contact ‚Üí Name
* Student ‚Üí Roll number
* Dictionary book ‚Üí Word



### Rules of Keys:

* Must be **unique**
* Must be **immutable** (cannot change)

### ‚úÖ Valid Keys:

```python
"name"      # string
10          # integer
(1, 2)      # tuple
```

### ‚ùå Invalid Keys:

```python
[1, 2]      # list
{1, 2}      # set
{"a": 1}    # dictionary
```

‚ùå These are mutable ‚Üí not allowed as keys.


##  What is a VALUE? üì¶


A **value** is the **actual data** stored in the dictionary.

### Rules of Values:

* Can be **any data type**
* Can be **mutable or immutable**
* Can be duplicated
* Can be changed anytime

### Examples:

```python
"Anisha"                 # string
24                       # integer
[80, 90, 85]             # list
{"city": "Bangalore"}    # dictionary
```

Q1. Write a program that asks user to enter the 5 product names and prices. Store all of these in a dictionary with keys are product names and prices are values. Then allow the user to input the product name and print the corresponding price of the product.

````python
#Try it here
#Q1 Solution

#Number of times asks the user to input the information
count = 5

#empty dictionary
products = {}

#Ask the user for product names and corresponding price, then add it to the dictionary
for i in range(count):
    product_name = input("Enter the product name\n")
    price = eval(input("Enter the product price\n"))
    products[product_name] = price


print("\n")
product_name = input("Enter the product name\n")
print("The corresponding price is ", products[product_name])
````

# Answer

````python
#Q1 Solution

#Number of times asks the user to input the information
count = 5

#empty dictionary
products = {}

#Ask the user for product names and corresponding price, then add it to the dictionary
for i in range(count):
    product_name = input("Enter the product name\n")
    price = eval(input("Enter the product price\n"))
    products[product_name] = price


print("\n")
product_name = input("Enter the product name\n")
print("The corresponding price is ", products[product_name])
````

````python
#Q2 Solution
months = {'January':31, 'February':28, 'March':31, 'April':30, "May":31, 'June':30, 'July':31, "August":31, 'September':30,
          'October':31, 'November':30, 'December':31}

#(a) Ask the user to input a month name and use the dictionary to tell how many days are in that month.
month_name = input("Enter the month\n")
print(month_name, " has ", months[month_name], " days.")

#(b) Print out all the keys in alphabetical order.
print("\nThe keys in alphabetical order : \n")
for key in sorted(months.keys()):
    print(key)
````

````python
#Q3 Solution

#Number of times asks the user to input the information
count = 2

#empty dictionary
teams = {}

#Ask the user for team names and corresponding wins & losses, then add it to the dictionary
for i in range(count):
    team_name = input("Enter the team name\n")
    wins = int(input("Enter the wins\n"))
    losses = int(input("Enter the losses\n"))
    teams[team_name] = [wins, losses]


print("\n")
team_name = input("Enter the team name\n")
print("The winning percentage is ", teams[team_name][0] / (teams[team_name][0] + teams[team_name][1]))

print("\n")
winning_list = []
for value in teams.values():
    winning_list.append(value[0])

print('The winning list is ', winning_list)
````
